**Q1** *Use the wine.csv dataset to realize the Principal Component Analysis technique*

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Wine Dataset from sklearn
wine_data = load_wine()
X = wine_data.data  # Features
y = wine_data.target  # Target classes
feature_names = wine_data.feature_names  # Feature names
```

```python
# Convert to DataFrame for easier manipulation
wine_df = pd.DataFrame(X, columns=feature_names)
wine_df['Target'] = y
```

```python
# Standardize the Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 components
X_pca = pca.fit_transform(X_scaled)
```

```python
# Combine PCA results with target labels for visualization
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Target'] = y
```

```python
# Visualize the Principal Components
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=pca_df, 
    x='PC1', 
    y='PC2', 
    hue='Target', 
    palette='Set1', 
    s=100
)
plt.title('PCA of Wine Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Target')
plt.grid(True)
plt.show()
```

```python
# Explained Variance
explained_variance = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance}")
print(f"Cumulative Explained Variance: {np.cumsum(explained_variance)}")
```
